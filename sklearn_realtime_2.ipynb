```python
from sklearn.externals import joblib
from confluent_kafka import Consumer, Producer, KafkaError
import json
```

```python
mnb_pipeline = joblib.load('nmb_pipeline.pkl')
```

```python
conf_c = {'bootstrap.servers': "localhost:9092",
            'group.id': 'predictor-p',
            'default.topic.config': {'acks': 'all'}
    }
```

```python
topic_c = 'test'
c = Consumer(conf_c)
c.subscribe([topic_c])
```
